```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 计算每个类别的电影数量
category_counts = df_film['category_id'].value_counts()

# 绘制条形图1
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
#sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Number of Films by Category ID')
plt.xlabel('Category ID')
plt.ylabel('Number of Films')
plt.show()
```

```python
# 绘制条形图2
plt.figure(figsize=(12, 6))
#category_counts.plot(kind='bar')
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Number of Films by Category Index')
plt.xlabel('Category Index')
plt.ylabel('Number of Films')
plt.show()
```

```python
# 根据分析'rental_duration' 是租赁时长的列，'category_id' 是电影类别的列
# 首先，按类别分组计算平均租赁时长
category_rental_duration = df_film.groupby('category_id')['rental_duration'].mean()

# 绘制箱型图，显示各个类别的租赁时长分布和平均值
plt.figure(figsize=(12, 6))
sns.boxplot(x='category_id', y='rental_duration', data=df_film)
plt.title('Average Rental Duration by Category with Variance')
plt.xlabel('Category ID')
plt.ylabel('Rental Duration')
plt.show()
```
